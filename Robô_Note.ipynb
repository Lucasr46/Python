```python
!pip install pyautogui

import pyautogui
import pyperclip
import time

pyautogui.PAUSE = 1


# pyautogui.click -> clicar
# pyautogui.press -> apertar 1 tecla
# pyautogui.hotkey -> conjunto de teclas
# pyautogui.write -> escreve um texto

# Passo 1: Entrar no sistema da empresa (no nosso caso é o link do drive)

pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(1)

# Passo 2: Navegar no sistema e encontrar a base de vendas (entrar na pasta exportar)

pyautogui.click(x=541, y=402, clicks=2)

# Passo 3: Fazer o download da base de vendas

time.sleep(1)
pyautogui.click(x=598, y=451)
time.sleep(1)
pyautogui.click(x=1618, y=235)
time.sleep(1)
pyautogui.click(x=1313, y=829)

# Passo 4: Importar a base de vendas pro Python -> utilizando Pandas

import pandas as pd

tabela = pd.read_excel(r"C:\Users\lucasr44\Downloads\Vendas - Dez.xlsx")
display (tabela)

# Passo 5: Calcular os indicadores da empresa

faturamento = tabela["Valor Final"].sum()
print (faturamento)
quantidade = tabela["Quantidade"].sum()
print(quantidade)

# Passo 6: Enviar um e-mail para a diretoria com os indicadores de venda

# abrir aba
time.sleep(1)
pyautogui.hotkey("ctrl","t")

# entrar no e-mail
pyperclip.copy("https://mail.google.com/mail/u/0/#inbox")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(1)

# clicar no botão escrever
pyautogui.click(x=146, y=248)

# preencher informações do e-mail
time.sleep(1)

pyautogui.write("lucaslhramos13@gmail.com")
pyautogui.press("tab") # selecionar e-mail

pyautogui.press("tab") # pular para o campo assunto
time.sleep(1)
pyperclip.copy("Relatório de Vendas")
pyautogui.hotkey("ctrl","v")

pyautogui.press("tab") # pular para o campo de corpo do e-mail

texto = f"""
Prezados,

Segue relatório de vendas.
Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {quantidade:,}
Em caso de dúvidas, fico à disposição
At. te,
Lucas Henrique
"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl","v")

#enviar o e-mail


```
